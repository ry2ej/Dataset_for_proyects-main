### Ejercicio 2. 

Una empresa química fabrica dos aditivos que permiten la elaboración de dos detergentes. Uno es utilizado especialmente
para la limpieza de artículos de vestuario y el otro para el lavado de utensilios de cocina. Para producir los dos aditivos se requiere mezclar tres materiales químicos, tal como se indica en siguiente tabla:

| Materiales | Aditivo Limpieza Vestuarios | Aditivo Utencilios Cocina |
| -- | -- | -- |
| Material A | .2 | .65|
| Material B | No contiene material B| .15 |
| Material c | .4 | .3 |

Para llevar a cabo la producción, se dispone de 14 toneladas del material A, 3 del material B y 12 del material C. Además, el gerente analizó las cifras de producción, y determinó una utilidad de S/37 por cada tonelada que se produzca a base de aditivo utilizado como detergente para limpieza de artículos de vestuario,y S/48 por cada tonelada producida de base de aditivo utilizado como detergente para limpieza de utensilios. Formule un problema de programación lineal (PL) que determine la cantidad de aditivo que es conveniente producir de acuerdo a su uso, a fin de maximizar la ganancia total.

Use **scipy** para resolver el problema de PL




**Solucion:**

    Variables: 
    
    Vestuario(V), Cocina(C)
    
    
   **Max** : 37*V + 48*C
   
   **Subjeto a** :
   
          0.2*V + 0.65*C <= 14
          0*V + 0.15*C <= 3
          0.4*V + 0.3*C <= 12
          
   **Min** : -37*V - 48*C
   
   **Subjeto a** :
   
          0.2*V + 0.65*C <= 14
          0*V + 0.15*C <= 3
          0.4*V + 0.3*C <= 12


```python
from scipy.optimize import linprog

# coeficientes de la funcion objetivo
c = [-37, -48]
# Matriz de restricciones de desigualdad 
A = [[0.2,0.65], [0,0.15], [0.4,0.3]]
# Vector de restricciones de desigualdades
b = [14,3,12]
# Calculo del optimo 
res = linprog(c, A_ub = A, b_ub = b)
res
```

```python
print("El valor optimo del problema es:  S/.", -res.fun)
print("Valor de las varibales que maximizan la solucion son: ")

variables= ["V", "C"]

arr = res.x
lstX = []

for x in arr:
    lstX.append(x)


for i in range(0,2):
    print(variables[i], " = ", lstX[i])
```

```python
import numpy as np
import matplotlib.pyplot as plt
V=np.linspace(2,20, 100) 

res1 = []
res2 = []
res3 = []

for v in V:
    res1.append((14-0.2*v)/0.65)
    res2.append((3-0*v)/0.15)
    res3.append((12-0.4*v)/0.3)

fig, ax = plt.subplots(figsize=(15,7))
ax.plot(V, res1, label='0.2*V + 0.65*C = 14')
ax.plot(V, res2, label='0*V + 0.15*C = 3')
ax.plot(V, res3, label='0.4*V + 0.3*C = 12')
ax.plot(17.99999995737966,15.999999958698274, marker="o", color="red", label='Solucion optima')
plt.xlabel('Aditivo Limpieza Vestuarios')
plt.ylabel('Aditivo Utencilios Cocina')
plt.legend()
plt.plot()
plt.grid()
```

Respaldando por la grafica podemos decir que la salucion encontrada por scipy es la correcta.

###  Ejercicio 3. 

Resuelva el siguiente problema de programación lineal :
    
   **Max** : 17,1667x + 25,8667y
        
   **Sujeto a** :
            
        13x + 19y ≤ 2400
        20x + 29y ≤ 2100
        x ≥ 10
        x, y ≥ 0
        
Use **pulp** para resolver el problema de PL

```python
import pulp
from pulp import LpProblem,LpMaximize,LpVariable,LpStatus,lpSum

model = LpProblem(name = "small_problem", sense = "LpMaximize")

x = LpVariable(name = "x", lowBound = 0)
y = LpVariable(name = "y", lowBound = 0)

model += (13*x + 19*y <= 2400, "prim_rest")
model += (20*x + 29*y <= 2100, "seg_rest")
model += (x + 0*y >= 10, "terc_rest")

obj_func =  17.1667*x + 25.8667*y

model += obj_func
model
```

```python
status = model.solve()
bool(status)
```

```python
print(LpStatus[model.status])
print("El valor optimo del problema es: ", model.objective.value())

print("Valor de las varibales son: ")
for var in model.variables():
    print(var, " = ",  var.value())
```

```python
import numpy as np
import matplotlib.pyplot as plt
X=np.linspace(10,80, 100) 

res1 = []
res2 = []
res3 = []

for x in X:
    res1.append((2400-13*x)/19)
    res2.append((2100-20*x)/29)
    res3.append(10)
    
fig, ax = plt.subplots(figsize=(15,7))
ax.plot(X, res1, label='13x + 19y = 2400')
ax.plot(X, res2, label='20x + 29y = 2100')
ax.plot(10.0,65.5172, marker="o", color="red", label='Solucion optima')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.plot()
plt.grid()
```

Del grafico podemos observar que la solucion dada por pulp es la correcta dado que seria la que maximize la el problema PL.

### Ejercicio 4. 

Resuelva el siguiente problema de programación lineal :

   **Min** : 4a + 5b + 6c
    
   **Sujeto a** :
   
            a + b ≥ 11
            a − b ≤ 5
            c − a − b = 0
            7a ≥ 35 − 12b
            a, b, c ≥ 0
        
Use **docplex** para resolver el problema de PL

```python
import docplex
from docplex.mp.model import Model

m = Model(name = "ThreeVars")

a = m.continuous_var(name = "a", lb = 0 )
b = m.continuous_var(name = "b", lb = 0 )
c = m.continuous_var(name = "c", lb = 0 )

r1 = m.add_constraint(a + b >= 11, ctname = "restri1")
r2 = m.add_constraint(a - b <= 5 , ctname = "restri2")
r3 = m.add_constraint(c - a - b == 0 , ctname = "restri3")
r4 = m.add_constraint(7*a +12*b >= 35 , ctname = "restri4")

m.set_objective("min", 4*a + 5*b + 6*c)

m.print_information()
```

```python
m.solve()
```
